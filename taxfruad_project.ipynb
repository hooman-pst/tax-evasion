<h2 style="color:red; font-family:Nazanin;" dir="rtl">1- هزینه سفر ها(total_trip_cost)</h2>


<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
برای محاسبه هزینه هر نوع سفر، می‌توان مراحل زیر را دنبال کرد:

1. **سفرهای هوایی زیارتی:** هزینه بلیط هواپیما، اقامت و غذا در مقصدهای زیارتی مانند مکه یا مشهد به‌طور میانگین محاسبه می‌شود.
  
2. **سفرهای هوایی غیرزیارتی:** هزینه بلیط، اقامت، غذا و تفریحات در مقصدهای تفریحی یا کاری در نظر گرفته می‌شود.

3. **سفرهای غیرهوایی زیارتی:** هزینه حمل و نقل (اتوبوس، قطار)، اقامت، غذا و هزینه‌های مذهبی محاسبه می‌شود.

4. **سفرهای غیرهوایی غیرزیارتی:** هزینه حمل و نقل، اقامت، غذا و تفریحات بر اساس مقصد تعیین می‌شوند. 

این روش‌ها به تخمین دقیق هزینه‌ها و تحلیل‌های مالی 
کمک می‌کنند.</p> 
د. 
</p>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
تمامی هزینه ها بر اساس تور های سایت های معتبر مانند علی بابا و اطلاعات خبرگزاری ها و همچنین قیمت بلیط های هواپیما و قطار و هزینه های ماشین شخصی بدست اورده شده همچنین از AI هم کمک گرفته شده

<a id = "parent"></a>
<h1 style="color:blue; font-family:Nazanin;" dir="rtl"> ثروت خانوار ها(سرپرست خانواده) </h1>

```python
#جدا کردن فیچر های مورد استفاده برای مصور سازی داده ها
parent_features = df.groupby('Parent_Id').agg({
    'Age': 'first',
    'GenderId': 'first',
    'SabteAhval_provincename': 'first',
    'SabteAhval_countyname': 'first',
    'Decile': 'first',
    'Percentile': 'first',
    'Yaraneh': 'first'
}).reset_index()
```

```python
parent_features
```

```python
df['Age'] = df.Age.astype('str')
df['id'] = df.id.astype('str')
df['GenderId'] = df.GenderId.astype('str')
```

```python
# پیمایش بر روی سرپرست های حانواده
df_family = df.pivot_table(index='Parent_Id',
                           values= ['id', 'Age','GenderId', 'Servat'],  
                           aggfunc= {'id':
                                     [np.count_nonzero, lambda x:','.join(x)],
                                     'Age':
                                     [lambda x:','.join(x)
                                      #, lambda x:set(x)
                                     ],
                                     'GenderId':
                                     [lambda x:','.join(x)],
                                     'Servat' : "sum"
                                     
                                    }

                          ).reset_index()
```

```python
df_family
```

```python
df_family.columns = ['Parent_Id', 'Age_family', 'GenderId_family', 'Servat', 'id_family', 'family_member']
```

```python
df_family
```

<ul style="color:red;font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    <li>پیوند دو داده جدا شده به یکدیگر </li>
</ul>
 </p> 

```python
df_family = df_family.merge(parent_features, on='Parent_Id', how='left')
```

```python
df_family
```

<a id = "Servat_Decile"></a>
<h1 style="color:blue; font-family:Nazanin;" dir="rtl">  ایجاد ستون 'Servat_Decile' </h1>

```python
percentile_80 = df_family['Servat'].quantile(0.8)
percentile_90 = df_family['Servat'].quantile(0.9)
```

```python
def servat_category(row):
    if row['Servat'] > percentile_90:
        return 'very high wealth'
    elif row['Servat'] > percentile_80:
        return 'High Wealth'
    else:
        return 'None'
        
```

```python
df_family['Servat_Decile'] = df_family.apply(servat_category, axis = 1) 
```

```python
df_family.head()
```

<a id = "Tax_fruad"></a>
<h1 style="color:blue; font-family:Nazanin;" dir="rtl">  ایجاد ستون 'Tax_fruad' </h1>

```python
df_family['Tax_fruad'] = ((df_family['Yaraneh'] == True) & (df_family['Servat_Decile'].isin(['High Wealth', 'Very High Wealth']))).astype('str')
```

```python
df_family['Tax_fruad'] = df_family['Tax_fruad'].map({'True': True, 'False': False})
```

```python
df_family['Tax_fruad'] = df_family['Tax_fruad'].astype(int)
```

### برطرف کردن باگ نخواندن استانها به زبان مرود استفاده در دیتاست(فارسی)

```python
import arabic_reshaper
import bidi.algorithm
from matplotlib import font_manager

def text_fa(x):
    x = arabic_reshaper.reshape(x)
    x = bidi.algorithm.get_display(x)
    return x


font = {'family' : 'B Mitra', 'size' : 21}
plt.rc('font', **font)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

# نمودار میله‌ای
plt.figure(figsize=(12, 8))
 
sns.countplot(y='SabteAhval_provincename', data=tax_evasion_df, order=tax_evasion_df['SabteAhval_provincename'].value_counts().index)

# اعمال تابع text_fa 
plt.yticks(ticks=range(len(tax_evasion_df['SabteAhval_provincename'].value_counts())), labels=[text_fa(x) for x in tax_evasion_df['SabteAhval_provincename'].value_counts().index])

plt.title(text_fa('فرار مالیاتی بر اساس استان'))
plt.xlabel(text_fa('تعداد فرارکنندگان مالیاتی'))
plt.ylabel(text_fa('استان'))
plt.show()
```
